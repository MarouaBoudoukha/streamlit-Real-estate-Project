```python
#Drop columns that have a rate of missing values greater than 90%
df_final_full20 = df_full20.drop(['numero_disposition',
                            'ancien_code_commune',
                            'adresse_suffixe',
                            'ancien_nom_commune',
                            'ancien_id_parcelle',
                            'numero_volume',
                            'lot1_numero',
                            'lot1_surface_carrez',
                            'lot2_numero',
                            'lot2_surface_carrez',
                            'lot3_numero',
                            'lot3_surface_carrez',
                            'lot4_numero',
                            'lot4_surface_carrez',
                            'lot5_numero',
                            'lot5_surface_carrez',
                            'code_nature_culture',
                            'nature_culture',
                            'code_nature_culture_speciale',
                            'nature_culture_speciale',
                            'code_commune'],axis= 1)
df_final_full20 
```

```python
#Vérifier le taux des données manquantes dans la base finale
df_final_full20.isnull().sum()/len(df_full20)*100
```

```python

df_final_full20.info()
```

```python

#Changer le types de variables (savoir la distribution des variables qualitatives et quantitatives afin de faire la correlation)
df_final_full20['code_postal'] = df_final_full20['code_postal'].astype(float)
df_final_full20['code_type_local'] = df_final_full20['code_type_local'].astype('object')
df_final_full20['type_local'] = df_final_full20['type_local'].astype('object')
df_final_full20['nombre_pieces_principales'] = df_final_full20['nombre_pieces_principales'].astype('object')
df_final_full20['code_departement'] = df_final_full20['code_departement'].astype('object')
#df_final_full20['code_commune'] = df_final_full20['code_commune'].type()
df_final_full20['nombre_lots'] = df_final_full20['nombre_lots'].astype('float64')
```

```python
df_final_full20.info()
```

```python
df_final_full20['type_local'].value_counts(dropna = False, normalize = True)
```

```python
#Traiter les données manquantes
# L'imputation des données manquantes
df_final_full20['adresse_numero'] = df_final_full20['adresse_numero'].fillna("None")
df_final_full20['adresse_nom_voie'] = df_final_full20['adresse_nom_voie'].fillna("None")
df_final_full20['adresse_code_voie'] = df_final_full20['adresse_code_voie'].fillna("None")
df_final_full20['code_type_local'] = df_final_full20['code_type_local'].fillna("None")
df_final_full20['type_local'] = df_final_full20['type_local'].fillna("None")
df_final_full20['valeur_fonciere'] = df_final_full20['valeur_fonciere'].transform(lambda x: x.fillna(x.mean()))
df_final_full20['surface_reelle_bati'] = df_final_full20['surface_reelle_bati'].fillna(0)
df_final_full20['nombre_pieces_principales'] = df_final_full20['nombre_pieces_principales'].fillna(0)
df_final_full20['longitude'] = df_final_full20['longitude'].fillna(0)
df_final_full20['latitude'] = df_final_full20['latitude'].fillna(0)
df_final_full20['surface_reelle_bati'] = df_final_full20['surface_reelle_bati'].fillna(0)
df_final_full20['surface_terrain'] = df_final_full20['surface_terrain'].fillna(0)
df_final_full20['latitude']=pd.to_numeric(df_final_full20['latitude']) 
df_final_full20['longitude']=pd.to_numeric(df_final_full20['longitude'])

```
